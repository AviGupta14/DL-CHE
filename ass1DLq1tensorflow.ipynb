```python
train_data1 = pd.read_excel("/kaggle/input/po2121/Folds5x2_pp.xlsx",sheet_name="Sheet1")
train_data2 = pd.read_excel("/kaggle/input/po2121/Folds5x2_pp.xlsx",sheet_name="Sheet2")
train_data3 = pd.read_excel("/kaggle/input/po2121/Folds5x2_pp.xlsx",sheet_name="Sheet3")
frames = [train_data1,train_data2,train_data3]
  
train_data = pd.concat(frames)
train_data.head()
```

```python
train_data
```

```python
test_data = pd.read_excel("/kaggle/input/po2121/Folds5x2_pp.xlsx", sheet_name="Sheet5")
test_data.head()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import MeanAbsoluteError
from sklearn.ensemble import RandomForestClassifier

y = train_data["PE"]

features = ["AT", "V", "AP", "RH"]
X = pd.get_dummies(train_data[features])

X_test = pd.get_dummies(test_data[features])
model = Sequential(
    [ 
        Dense(25, activation = 'relu'),
        
        Dense(10, activation = 'relu'),
        
        Dense(1, activation = 'relu')# < softmax activation here
    ]
)
model.compile(
    loss=tf.keras.losses.MeanAbsoluteError(),
    optimizer=tf.keras.optimizers.Adam(0.001),
)

model.fit(
    X,y,
    epochs=1000
)
predictions = model.predict(X_test)
```
