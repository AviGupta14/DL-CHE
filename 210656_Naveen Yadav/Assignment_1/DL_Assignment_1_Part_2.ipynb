```python
#Fill the null values in age column with the mean
df['Age'].fillna(df['Age'].mean(),inplace=True)
df.isnull().sum()
```

```python
#Since the Cabin column does not contain numerical values and the number of null cells is very large, we can frop this column

df.drop('Cabin',axis=1,inplace=True)
df.head()
```

```python
gender=pd.get_dummies(df['Sex'],drop_first=True)
df['Gender']=gender
df.drop('Sex',axis=1,inplace=True)
df.head()
```

```python
df.drop(['Name','Ticket','Embarked','PassengerId'],axis=1,inplace=True)
df.head()
```

```python
input=['Pclass','Age',	'SibSp',	'Parch',	'Fare',	'Gender']
X=torch.tensor(df[input].values,dtype=torch.float32)
output=['Survived']
y=torch.tensor(df[output].values,dtype=torch.float32)
```

```python
X.shape
```

```python
class LR(nn.Module):

  def __init__(self,input_features):
    super(LR,self).__init__()
    self.linear=nn.Linear(input_features,1)

  def forward(self,X):
    y_pred=torch.sigmoid(self.linear(X))
    return y_pred

model=LR(6)
```

```python
criterion=nn.BCELoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.05)

for epoch in range(800):
  y_pred=model(X)
  loss=criterion(y_pred,y)

  loss.backward()
  optimizer.step()
  optimizer.zero_grad()

  if((epoch+1)%100==0):
    print("LOSS= ",loss)

```

```python
with torch.no_grad():
  y_predicted=model(X)
  y_prediction=y_predicted.round()
  accuracy=y_prediction.eq(y).sum()/float(y.shape[0])

  print("Accuracy= " ,accuracy)
```
