```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
data = pd.read_csv("E:\downloads\data.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
stdPE = 17.066995
meanPE = 454.365009
```

```python
cr = data.corr()
cr
```

```python
data=(data-data.mean())/data.std()
data.head()
```

```python
X = data.iloc[:,0:4]
Y = data.iloc[:,4]
X = torch.tensor(X.values)
Y = torch.tensor(Y.values)
```

```python
X_train_val,X_test,Y_train_val,Y_test = train_test_split(X,Y,test_size=0.1,random_state=42)
X_train,X_valid,Y_train,Y_valid = train_test_split(X_train_val,Y_train_val,test_size=0.25,random_state=42)
X_train = X_train.float()
Y_train = Y_train.float()
X_valid = X_valid.float()
Y_valid = Y_valid.float()
X_test = X_test.float()
Y_test = Y_test.float()
```

```python
device = "cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu"
```

# Model1


```python
model1=nn.Sequential(
    nn.Linear(4,64,dtype=None),
    nn.ReLU(),
    nn.Linear(64,32,dtype=None),
    nn.ReLU(),
    nn.Linear(32,1,dtype=None),
)
model1.to(device)
```

```python
ypred = model1(X_train[1])
print(ypred)
ypred=torch.squeeze(ypred)
print(ypred)
print(Y_train[1])
print(Y_train[1].shape)
```

```python
optimizer=torch.optim.Adam(model1.parameters(),lr=0.01)
for epoch in range(5):
  totalLoss=0
  for i in range(len(X_train)):
    ypred=torch.squeeze(model1(X_train[i]))
    loss=F.mse_loss(ypred,Y_train[i])
    totalLoss+=loss.item()

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
  print("Total Loss= ", totalLoss)
   
```

# Prediction on validation data



```python
prediction=[]
true_val=[]
total_Loss=0
for i in range(len(X_valid)):
    pred=model1(X_valid[i])
    pred=pred*stdPE + meanPE
    true_value=Y_valid[i]*stdPE +meanPE
    Loss = F.mse_loss(torch.squeeze(pred),true_value)
    total_Loss+=Loss
    prediction.append(pred.tolist())
    true_val.append(true_value.item()) 
print("Tatal loss",total_Loss)    
plt.scatter(true_val,prediction)
plt.xlabel('Actual PE')
plt.ylabel('Predicted PE')

plt.plot([415,505],[415,505])
plt.xlim(415,505)
plt.ylim(415,505)
    
```

# Model2

```python
model2=nn.Sequential(
    nn.Linear(4,128,dtype=None),
    nn.ReLU(),
    nn.Linear(128,64,dtype=None),
    nn.ReLU(),
    nn.Linear(64,32,dtype=None),
    nn.ReLU(),
    nn.Linear(32,1,dtype=None),
)
model2.to(device)
```

```python
ypred = model2(X_train[1])
print(ypred)
ypred=torch.squeeze(ypred)
print(ypred)
print(Y_train[1])
print(Y_train[1].shape)
```

```python
optimizer=torch.optim.Adam(model2.parameters(),lr=0.01)
for epoch in range(5):
  totalLoss=0
  for i in range(len(X_train)):
    ypred=torch.squeeze(model2(X_train[i]))
    loss=F.mse_loss(ypred,Y_train[i])
    totalLoss+=loss.item()

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
  print("Total Loss= ", totalLoss)
   
```

```python
prediction=[]
true_val=[]
total_Loss=0
for i in range(len(X_valid)):
    pred=model2(X_valid[i])
    pred=pred*stdPE + meanPE
    true_value=Y_valid[i]*stdPE +meanPE
    Loss = F.mse_loss(torch.squeeze(pred),true_value)
    total_Loss+=Loss
    prediction.append(pred.tolist())
    true_val.append(true_value.item()) 
print("Tatal loss",total_Loss)    
plt.scatter(true_val,prediction)
plt.xlabel('Actual PE')
plt.ylabel('Predicted PE')

plt.plot([415,505],[415,505])
plt.xlim(415,505)
plt.ylim(415,505)
    
```

# So based on the validation loss, Model 1 is more accurate
