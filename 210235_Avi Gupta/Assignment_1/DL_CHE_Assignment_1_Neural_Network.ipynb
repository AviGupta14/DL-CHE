```python
import torch
from torch import nn
from torch.utils.data import DataLoader, TensorDataset
from torchvision import datasets
from torchvision.transforms import ToTensor
import numpy as np
import pandas as pd
import tensorflow as tf
from torch.utils.data import Dataset
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import torch.nn.functional as F
```

```python
xls = pd.ExcelFile("data.xlsx")
df1 = xls.parse('Sheet1')
df2 = xls.parse('Sheet2')
df3 = xls.parse('Sheet3')
df4 = xls.parse('Sheet4')
df5 = xls.parse('Sheet5')
```

```python
frames = [df1, df2, df3]

df_train = pd.concat(frames)
```

```python
frames = [df4, df5]

df_test = pd.concat(frames)
```

```python
x_train=df_train.iloc[:,0:4].values
y_train=df_train.iloc[:,4].values
```

```python
x_train = torch.tensor(x_train, dtype=torch.float)
y_train = torch.tensor(y_train, dtype=torch.float)
```

```python
x_train.shape, y_train.shape
```

```python
x_test=df_test.iloc[:,0:4].values
y_test=df_test.iloc[:,4].values
```

```python
x_test = torch.tensor(x_test, dtype=torch.float)
y_test = torch.tensor(y_test, dtype=torch.float)
```

```python
x_test.shape, y_test.shape
```

```python
train_data = TensorDataset(x_train, y_train)
test_data = TensorDataset(x_test, y_test)
```

```python
train_dataloader = DataLoader(train_data, shuffle=True, batch_size=64)
```

```python
test_dataloader = DataLoader(test_data, batch_size=64)
```

```python
print("Training data batches:")
for X, y in train_dataloader:
    print(X.shape, y.shape)
    
print("\nTest data batches:")
for X, y in test_dataloader:
    print(X.dtype, y.dtype)
```

```python
# Get cpu or gpu device for training.
device = "cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu"
print(f"Using {device} device")

# Define model
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(4, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 1)
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits

model = NeuralNetwork().to(device)
print(model)
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    model.train()
    for batch, (X, y) in enumerate(dataloader):
        y = y.unsqueeze(1)
        X, y = X.to(device), y.to(device)

        # Compute prediction error
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), (batch + 1) * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
```

```python
def test(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()
    test_loss, correct = 0, 0
    with torch.no_grad():
        for X, y in dataloader:
            y = y.unsqueeze(1)
            X, y = X.to(device), y.to(device)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

```python
epochs = 10
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_dataloader, model, loss_fn, optimizer)
    test(test_dataloader, model, loss_fn)
print("Done!")
```
